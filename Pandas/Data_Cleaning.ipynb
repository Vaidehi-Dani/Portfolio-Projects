```python
import pandas as pd 
```

```python
df = pd.read_excel(r"C:\Users\vaide\Downloads\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
print(df.columns)
```

```python
df = df.drop(columns="Not_Useful_Column", errors='ignore')
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("abc./_")
df["Last_Name"]
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
df["Phone_Number"]
```

```python
# Clean 'N/a' or missing values
df["Address"] = df["Address"].replace("N/a", "").fillna("")

# Split the address into columns based on commas
df_split = df["Address"].str.split(",", expand=True)

# Rename the columns for clarity
df_split.columns = ["Street", "City", "State"]

df.drop(["Street", "City", "State"], axis=1, errors="ignore")

# Add the split columns back to the original DataFrame, but only if they are not already present
df = pd.concat([df.drop(["Street", "City", "State"], axis=1, errors="ignore"), df_split], axis=1)

# Display the updated DataFrame
print(df)
```

```python
# Drop the Address column
df = df.drop("Address", axis=1)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a', '')
df = df.replace('NaN', '')
df = df.fillna('')
df
```

```python
# Drop rows where 'Do_Not_Contact' is 'Yes'
df = df[df['Do_Not_Contact'] != 'Yes']

df
```

```python
# Drop rows where 'Phone_Number' is null (NaN)
df = df.dropna(subset=['Phone_Number'])
df
```

```python
df["Phone_Number"] 
```

```python
# Convert all values in 'Phone_Number' to strings, replacing NaN with empty strings
df['Phone_Number'] = df['Phone_Number'].astype(str).fillna('')

# Remove all non-numeric characters and standardize the format
df['Phone_Number'] = (
    df['Phone_Number']
    .str.replace(r'[^0-9]', '', regex=True)  # Remove non-numeric characters
    .str.slice(start=0, stop=10)            # Take the first 10 digits (if valid)
    .apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}" if len(x) == 10 else '')  # Format or set empty
)
df
```
